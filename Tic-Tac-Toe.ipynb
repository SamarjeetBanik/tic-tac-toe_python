# Tic-Tac-Toe Game (Without GUI)

```python
import random # a library that will help in generating random numbers.

"""
Flow of the program
----------------------
1. A menu showing the modes of the play.
2. There are two modes:
    a. Auto (User Vs Computer)
    b. Manual (User Vs Another User)
3. If someone chooses 3rd option from menu, he/she will be exited from the game.
4. After choosing the one of modes, the user have to input name and choose the mark (X or O).
5. If Auto mode:
    a. The computer/bot will make it's first move.
    b. Then the user will have to mention the index of the 3x3 board
    c. If the user quits, in the middle of the game, then as per rule, the BOT wins.
6. If Manual mode:
    a. The rival is a person, and had to input name.
    b. Then one-by-one, both will put their marks according to the index of the 3x3 board.
    c. If any one of the players/users quits  in the middle of the game, then the other person/user will win, as per rule.
7. After the full match is being played, as per the winning strategy, one will win and a winning note will be displayed.
8. In the end, the program would ask the user, if he/she wants to continue the whole game once again or not.
9. If yes, then the all steps from 1 will begin, else, the program terminates with a thank you note.
"""

# a function that displays the board and updates according to the program flow.
def board(player, rival, boardVal, mark, rivalMark):
    print("\t\t\t  _____________________________________")
    print("\t\t\t |           |            |            |")
    print("\t\t\t | "+boardVal[0]+"         |  "+boardVal[1]+"         |  "+boardVal[2]+"         |")
    print("\t\t\t |           |            |            |")
    print("\t\t\t  _____________________________________")
    print("\t\t\t |           |            |            |")
    print("\t\t\t | "+boardVal[3]+"         |  "+boardVal[4]+"         |  "+boardVal[5]+"         |")
    print("\t\t\t |           |            |            |")
    print("\t\t\t  _____________________________________")
    print("\t\t\t |           |            |            |")
    print("\t\t\t | "+boardVal[6]+ "         |  "+boardVal[7]+"         |  "+boardVal[8]+"         |")
    print("\t\t\t |           |            |            |")
    print("\t\t\t  _____________________________________")

# a function that checks and evaluate the winning strategy
def winner(boardVal, symbol):
    if((boardVal[0] == symbol and boardVal[1] == symbol and boardVal[2] == symbol) or #  1st row
        (boardVal[3] == symbol and boardVal[4] == symbol and boardVal[5] == symbol) or # 2nd row
        (boardVal[6] == symbol and boardVal[7] == symbol and boardVal[8] == symbol) or # 3rd row
        (boardVal[0] == symbol and boardVal[3] == symbol and boardVal[6] == symbol) or # 1st column
        (boardVal[1] == symbol and boardVal[4] == symbol and boardVal[7] == symbol) or # 2nd column
        (boardVal[2] == symbol and boardVal[5] == symbol and boardVal[8] == symbol) or # 3rd column
        (boardVal[0] == symbol and boardVal[4] == symbol and boardVal[8] == symbol) or # diagonal1 from left
        (boardVal[2] == symbol and boardVal[4] == symbol and boardVal[6] == symbol)): # diagonal2 from right
        return True
    else:
        return False

while(True): # infinite loop
    print('\n\t\t\t     TIC TAC TOE')
    print("--------------------------------------------------------------------\n")
    print("\t  MODE\n 1 - Auto (Vs Computer)\n 2 - Manual (Vs Person)\n 3 - Quit")
    ch = input("Enter Choice: ")
    
    if(ch == "3"):
        break
    while (ch != "1" and ch != "2"):
        print("Invalid Choice!")
        ch = input("Enter Choice: ")
        
    ch = ord(ch) - 48 # convert the input to int (in range from 0 to 9).
    player = input("Enter Your Name: ")
    if (ch == 1):
        rival = "BOT"
    else:
        rival = input("Enter 2nd Player's Name: ")
    boardVal = [" ", " ", " ", " ", " ", " ", " ", " ", " "]
    count = 0
    mark = input("Choose Your Symbol  (X , O): ").upper()
    while (mark != "X" and mark != "O"):
        mark = input("Please choose either 'X' or 'O': ").upper()
    if (mark == "X"):
        rivalMark = "O"
    else:
        rivalMark = "X"

    print("\t\t\t", player + "[" + mark + "]", "   vs   ", rival + "[" + rivalMark + "]")

    board(player, rival, boardVal, mark, rivalMark) # passing the values to the 'board' function as arguments to update the board.
    while (count < 9):
        if (count % 2 == 0):
            if (ch == 1):
                choice = random.randint(1, 9)
            else:
                print(rival, " Turn")
                choice = input("Enter Choice (or quit): ")
                if(choice == "Quit" or choice == "quit" or choice == "Q" or choice == "q"):
                    print("Congratulations! " + player + " Won! (" + rival + " left the game)")
                    break
                choice = ord(choice) - 48
                if(choice > 0 and choice < 10):
                    if (boardVal[choice - 1] == rivalMark or boardVal[choice - 1] == mark):
                        print("Invalid Choice")
                        continue
                else:
                    print("Invalid Choice")
                    continue

            if (boardVal[choice - 1] != rivalMark and boardVal[choice - 1] != mark): # checking whether the index enterd for marking, is filled with prior value or not.
                boardVal[choice - 1] = rivalMark
                count += 1
                print(rival, " Turn")
                board(player, rival, boardVal, mark, rivalMark)
                if (winner(boardVal, rivalMark)): # to check if the rival is winning or not
                    print("Better Luck Next Time! (" + rival + " Won!)")
                    count = 0
                    break

        else:
            print(player, " Turn")
            choice = input("Enter Choice (or quit): ")
            if(choice == "Quit" or choice == "quit" or choice == "Q" or choice == "q"):
                print("You left the game. (" + rival + " Won!)")
                break
            choice = ord(choice) - 48
            if (choice > 0 and choice < 10):
                if (boardVal[choice - 1] == rivalMark or boardVal[choice - 1] == mark): # checking whether the index enterd for marking, is filled with prior value or not.
                    print("Invalid Position")
                else:
                    boardVal[choice - 1] = mark
                    count += 1
                    board(player, rival, boardVal, mark, rivalMark)
                    if (winner(boardVal, mark)): # to check if the player is winning or not
                        print("Congratulations! " + player + " Won!")
                        break
            else:
                print("Invalid Choice")
    if (count > 8):
        print("Game Tied!")
    
    play = input("Do you want to continue? (Yes/No): ")
    if(play == "Yes" or play == "yes" or play == "Y" or play == "y"):
        continue
    else:
        print("--------------------------------X-----------------------------------")
        print("Thank You!")
        break
```
